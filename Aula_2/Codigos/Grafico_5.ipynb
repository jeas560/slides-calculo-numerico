```python
#Importando as bibliotecas utilizadas
from mpl_toolkits.axisartist.axislines import AxesZero
import numpy as np  #biblioteca numérica
import matplotlib.pyplot as plt

save_figures = True

fig = plt.figure()
ax = fig.add_subplot(axes_class = AxesZero)

for direction in ["xzero", "yzero"]:
    # adds arrows at the ends of each axis
    ax.axis[direction].set_axisline_style("-|>")

    # adds X and Y-axis from the origin
    ax.axis[direction].set_visible(True)

for direction in ["left", "right", "bottom", "top"]:
    # hides borders
    ax.axis[direction].set_visible(False)

x = np.arange(0.1, 0.55, 0.01)
y = 0.5 + np.sin(1.15*np.pi + 2.5*np.pi*x)

plt.yticks([ ])
plt.xticks([0.15, 0.27, 0.45], ["a", r"$\alpha$","b"])
plt.xlim([-0.01, 0.6])

plt.plot(x, y, label = r'$f^{\prime}(x) > 0, \forall \in [a,b]$')
plt.legend(framealpha=1, frameon = True, loc ='lower center')

ax.set_xlabel("x")
ax.set_ylabel("y")

if save_figures:
    plt.savefig('../Figuras/grafico_05a.pdf')
plt.show()
```

```python
#Importando as bibliotecas utilizadas
from mpl_toolkits.axisartist.axislines import AxesZero
import numpy as np  #biblioteca numérica
import matplotlib.pyplot as plt

save_figures = True

fig = plt.figure()
ax = fig.add_subplot(axes_class = AxesZero)

for direction in ["xzero", "yzero"]:
    # adds arrows at the ends of each axis
    ax.axis[direction].set_axisline_style("-|>")

    # adds X and Y-axis from the origin
    ax.axis[direction].set_visible(True)

for direction in ["left", "right", "bottom", "top"]:
    # hides borders
    ax.axis[direction].set_visible(False)

x = np.arange(0.05, 0.42, 0.005)
y = - x*np.sin(0.5*np.pi + 7.0*np.pi*x)

plt.yticks([ ])
plt.xticks([0.1, 0.37], ["a","b"])
plt.xlim([-0.01, 0.5])

plt.plot(x, y, label = '$f^{\prime}(x)$ não preserva o sinal')
plt.legend(framealpha=1, frameon = True, loc ='lower center')

ax.set_xlabel("x")
ax.set_ylabel("y")

if save_figures:
    plt.savefig('../Figuras/grafico_05c.pdf')
plt.show()
```
