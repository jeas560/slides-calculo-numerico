```python
#Importando as bibliotecas utilizadas
from mpl_toolkits.axisartist.axislines import AxesZero
import numpy as np  #biblioteca numérica
import matplotlib.pyplot as plt

save_figures = True

fig = plt.figure()
ax = fig.add_subplot(axes_class = AxesZero)

for direction in ["left", "bottom"]:
    # adds arrows at the ends of each axis
    ax.axis[direction].set_axisline_style("-|>")

for direction in ["right", "top"]:
    # hides borders
    ax.axis[direction].set_visible(False)

xv = np.array([-1, 0, 1])
x = np.linspace(-1.0, 1.0, 100)
L0 = (x - xv[1])*(x - xv[2])/((xv[0] - xv[1])*(xv[0] - xv[2]))
L1 = (x - xv[0])*(x - xv[2])/((xv[1] - xv[0])*(xv[1] - xv[2]))
L2 = (x - xv[0])*(x - xv[1])/((xv[2] - xv[0])*(xv[2] - xv[1]))

plt.plot(x, L0, label = r'$ L_{0} (x) $')
plt.plot(x, L1, label = r'$ L_{1} (x) $')
plt.plot(x, L2, label = r'$ L_{2} (x) $')

plt.yticks([0, 0.5, 1.0])
plt.xticks([-1, -0.5, 0, 0.50, 1])

plt.legend(framealpha=1, frameon = True, loc ='upper center')

ax.set_xlabel("x")
ax.set_ylabel("y")

if save_figures:
    plt.savefig('../Figuras/grafico_04.pdf')
plt.show()
```
