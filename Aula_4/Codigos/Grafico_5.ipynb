```python
#Importando as bibliotecas utilizadas
from mpl_toolkits.axisartist.axislines import AxesZero
import numpy as np  #biblioteca numérica
import matplotlib.pyplot as plt


#Definindo as funções Li(x)
def L0(x,xv):
    return (x - xv[1])*(x - xv[2])*(x - xv[3])/((xv[0] - xv[1])*(xv[0] - xv[2])*(xv[0] - xv[3]))

def L1(x,xv):
    return (x - xv[0])*(x - xv[2])*(x - xv[3])/((xv[1] - xv[0])*(xv[1] - xv[2])*(xv[1] - xv[3]))

def L2(x,xv):
    return (x - xv[0])*(x - xv[1])*(x - xv[3])/((xv[2] - xv[0])*(xv[2] - xv[1])*(xv[2] - xv[3]))

def L3(x,xv):
    return (x - xv[0])*(x - xv[1])*(x - xv[2])/((xv[3] - xv[0])*(xv[3] - xv[1])*(xv[3] - xv[2]))

save_figures = True

fig = plt.figure()
ax = fig.add_subplot(axes_class = AxesZero)

for direction in ["left", "bottom"]:
    # adds arrows at the ends of each axis
    ax.axis[direction].set_axisline_style("-|>")

for direction in ["right", "top"]:
    # hides borders
    ax.axis[direction].set_visible(False)

x1 = np.array([1, 1.10, 1.20, 1.3])
y1 = np.tan(x1)

x = np.linspace(1, 1.3, 100)
y2 = np.tan(x)
y3 = y1[0]*L0(x,x1) + y1[1]*L1(x,x1) + y1[2]*L2(x,x1) + y1[3]*L3(x,x1)

plt.plot(x, y2, label = '$ tan (x) $')
plt.plot(x, y3, '--', label = r'$ P_{1} (x) $')
plt.plot(x1, y1, 'ro')
plt.plot(1.15, np.tan(1.15), 'bo')
plt.yticks([1.6, 2.0, 2.4, 2.8, 3.2])
plt.xticks([1, 1.10, 1.15, 1.20, 1.3])

plt.legend(framealpha=1, frameon = True, loc ='upper center')

ax.set_xlabel("x")
ax.set_ylabel("y")

if save_figures:
    plt.savefig('../Figuras/grafico_05.pdf')
plt.show()
```
