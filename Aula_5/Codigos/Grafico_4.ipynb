```python
#Importando as bibliotecas utilizadas
from mpl_toolkits.axisartist.axislines import AxesZero
import numpy as np  #biblioteca numÃ©rica
import matplotlib.pyplot as plt

save_figures = True

fig = plt.figure()
ax = fig.add_subplot(axes_class = AxesZero)

for direction in ["xzero", "yzero"]:
    # adds arrows at the ends of each axis
    ax.axis[direction].set_axisline_style("-|>")

    # adds X and Y-axis from the origin
    ax.axis[direction].set_visible(True)

for direction in ["left", "right", "bottom", "top"]:
    # hides borders
    ax.axis[direction].set_visible(False)

x = np.array([0.3, 2.7, 4.5, 5.9, 7.8])
y = np.array([1.8, 1.9, 3.1, 3.9, 3.3])

x1 = np.array([0.3,7.8])
y1 = 1.74 + 0.2*x1

plt.plot(x1, y1, label = r"$g_{1}(x)$")
plt.plot([2.7,2.7], [1.9, 1.74 + 0.2*2.7], 'g-')
plt.plot([4.5,4.5], [3.1, 1.74 + 0.2*4.5], 'g-')
plt.plot([5.9,5.9], [3.9, 1.74 + 0.2*5.9], 'g-')

plt.plot(x, y,  'ro', label = r"$f(x)$")


plt.yticks([0,1,2,3,4,5])
plt.xticks([2,4,6,8])

plt.ylim([0,5.5])
plt.xlim([0,8])

plt.legend(framealpha=1, frameon = True, loc ='upper center')

ax.set_xlabel("x")
ax.set_ylabel("y")

if save_figures:
    plt.savefig('../Figuras/grafico_04.pdf')
plt.show()
```
