```python
#Importando as bibliotecas utilizadas
from mpl_toolkits.axisartist.axislines import AxesZero
import numpy as np  #biblioteca numérica
import matplotlib.pyplot as plt

save_figures = True

fig = plt.figure()
ax = fig.add_subplot(axes_class = AxesZero)

for direction in ["xzero", "yzero"]:
    # adds arrows at the ends of each axis
    ax.axis[direction].set_axisline_style("-|>")

    # adds X and Y-axis from the origin
    ax.axis[direction].set_visible(True)

for direction in ["left", "right", "bottom", "top"]:
    # hides borders
    ax.axis[direction].set_visible(False)

x = np.array([0, 0.25, 0.5, 0.75, 1.0])
y = np.array([1.0, 1.2840, 1.6487, 2.1170, 2.7183])

x1 = np.array([0,1.0])
y1 =  0.89968 + 1.70784*x1

plt.plot(x1, y1, label = r"$g(x)$")
plt.plot(x, y,  'ro')


plt.yticks([1,2,3])
plt.xticks([0, 0.25, 0.5, 0.75, 1.0])

plt.ylim([0,3])
plt.xlim([-0.05,1.05])

plt.legend(framealpha=1, frameon = True, loc ='upper center')

ax.set_xlabel("x")
ax.set_ylabel("y")

if save_figures:
    plt.savefig('../Figuras/grafico_05.pdf')
plt.show()
```
